Normalization and Data Cleaning

```python
import pandas as pd

path = "../archive/top_10000_1960-now.csv"

df = pd.read_csv(path)

columns_to_keep = ['Track URI', 'Track Name', 'Album URI']
df_track = df[columns_to_keep]

df_track['Track URI'] = df_track['Track URI'].str.strip()
df_track['Track Name'] = df_track['Track Name'].str.strip()
df_track['Album URI'] = df_track['Album URI'].str.strip()

df_track.replace("", pd.NA, inplace=True)

df_unique_tracks = df_track.drop_duplicates(subset='Track URI')

df_unique_tracks = df_unique_tracks.dropna(subset=['Track URI', 'Track Name', 'Album URI'])

print(df_unique_tracks)
```

```python
df_unique_tracks.to_csv('song.csv', index=False, encoding='utf-8')
```

```python
import psycopg2

conn = psycopg2.connect(
    dbname="musicdb",  
    user="postgres",   
    password="postgres", 
    host="localhost",   
    port="5432"     
)

cur = conn.cursor()
```

```python
create_table_query = """
CREATE TABLE artist (
    artistURI VARCHAR(300) PRIMARY KEY,
    name VARCHAR(150) NOT NULL
);
"""

cur.execute(create_table_query)
conn.commit()
```

```python
import pandas as pd

file_path = './artist.csv' 
df = pd.read_csv(file_path)

for index, row in df.iterrows():
    insert_query = """
    INSERT INTO artist (artistURI, name)
    VALUES (%s, %s)
    """
    cur.execute(insert_query, (row['Artist URI(s)'], row['Artist Name(s)']))

conn.commit()
```

```python
create_table_query = """
CREATE TABLE album (
    albumURI VARCHAR(300) PRIMARY KEY,
    title VARCHAR(200) NOT NULL,
    artist_uri VARCHAR(300),
    releaseDate DATE,
    CONSTRAINT fk_artist FOREIGN KEY (artist_uri) REFERENCES artist (artistURI) ON DELETE SET NULL
);
"""

cur.execute(create_table_query)
conn.commit() 
```

```python
import pandas as pd
import psycopg2
from datetime import datetime

df_albums = pd.read_csv("./album.csv")

def fix_date(date_str):
    try:
        return datetime.strptime(date_str, "%Y-%m-%d").date()
    except ValueError:
        try:
            return datetime.strptime(date_str, "%Y").date().replace(month=1, day=1)
        except ValueError:
            try:

                return datetime.strptime(date_str, "%Y-%m").date().replace(day=1)
            except ValueError:
                return None 

for index, row in df_albums.iterrows():
    try:
        artist_uri = row['Album Artist URI(s)']
        cur.execute("SELECT 1 FROM artist WHERE artistURI = %s", (artist_uri,))
        if cur.fetchone():
            release_date = fix_date(row['Album Release Date'])
            if release_date:
                insert_query = """
                INSERT INTO album (albumURI, title, artist_uri, releaseDate)
                VALUES (%s, %s, %s, %s)
                """
                cur.execute(insert_query, (row['Album URI'], row['Album Name'], artist_uri, release_date))
            else:
                print(f"Line {index}: invalide's date")
        else:
            release_date = fix_date(row['Album Release Date'])
            if release_date:
                insert_query = """
                INSERT INTO album (albumURI, title, artist_uri, releaseDate)
                VALUES (%s, %s, NULL, %s)
                """
                cur.execute(insert_query, (row['Album URI'], row['Album Name'], release_date))
            else:
                print(f"Line {index}: skipped")
    except psycopg2.Error as e:
        print(f"Error line {index}: {e}")

conn.commit()
```

```python
create_table_query = """
CREATE TABLE song (
    trackURI VARCHAR(300) PRIMARY KEY,
    title VARCHAR(200) NOT NULL,
    album_uri VARCHAR(300),
    FOREIGN KEY (album_uri) REFERENCES album (albumURI) ON DELETE SET NULL
);
"""

cur.execute(create_table_query)
conn.commit() 
```

```python
create_table_query = """
CREATE TABLE users (
    email VARCHAR(200) PRIMARY KEY NOT NULL,
    username VARCHAR(150) UNIQUE NOT NULL,
    gender VARCHAR(50) NOT NULL,
    birthday DATE NOT NULL,
    subscription_plan VARCHAR(50),
    FOREIGN KEY (subscription_plan) REFERENCES subscriptionPlan(name)
);
""" #user is not used because it is a reserved word in PostgreSQL

cur.execute(create_table_query)
conn.commit() 
```

```python
import pandas as pd
import random

subscription_plans = [
    ("Free", 0.00, "Free membership with advertising, limited to 30 songs per day"),
    ("Plus", 4.99, "Ad-free membership, limited to online listening only"),
    ("Premium", 9.99, "All incluse")
]

try:
    for plan in subscription_plans:
        cur.execute("""
            INSERT INTO subscriptionPlan (name, cost, description) 
            VALUES (%s, %s, %s) ON CONFLICT (name) DO NOTHING;
            """, plan
        )
    conn.commit()
except psycopg2.Error as e:
    print(f"Error: {e}")
```

```python
import pandas as pd
import psycopg2
import random

df_users = pd.read_csv("./user.csv")
available_plans = [plan[0] for plan in subscription_plans]

for index, row in df_users.iterrows():
    try:
        subscription_plan = random.choice(available_plans)
        insert_query = """
        INSERT INTO users (email, username, gender, birthday, subscription_plan)
        VALUES (%s, %s, %s, %s, %s)
        """
        cur.execute(insert_query, (row['Email'], row['Username'], row['Gender'], row['BirthDate'], subscription_plan))
    except psycopg2.Error as e:
        print(f"Error {index}: {e}")

conn.commit()
```

```python
create_table_query = """
CREATE TABLE listens (
    user_email VARCHAR(200) NOT NULL,
    song_uri VARCHAR(300) NOT NULL,
    rate INT,
    number_listen INT NOT NULL,
    PRIMARY KEY (user_email, song_uri),
    FOREIGN KEY (user_email) REFERENCES users(email) ON DELETE CASCADE,
    FOREIGN KEY (song_uri) REFERENCES song(trackURI) ON DELETE CASCADE
);
"""

cur.execute(create_table_query)
conn.commit() 
```

```python
import pandas as pd
import psycopg2
import random

users_df = pd.read_csv("./user.csv")
songs_df = pd.read_csv("./song.csv")

generated_pairs = set()

for _ in range(2000):
    #With just random there may be repetitions of associations
    while True:
        user_email = random.choice(users_df['Email'])
        song_uri = random.choice(songs_df['Track URI'])
        if (user_email, song_uri) not in generated_pairs:
            generated_pairs.add((user_email, song_uri))
            break
    rate = random.randint(1, 5) if random.random() > 0.5 else None 
    number_listen = random.randint(1, 100)

    try:
        insert_query = """
        INSERT INTO listens (user_email, song_uri, rate, number_listen)
        VALUES (%s, %s, %s, %s)
        """
        cur.execute(insert_query, (user_email, song_uri, rate, number_listen))
    except psycopg2.Error as e:
        print(f"Error: {e}")

conn.commit()
```

```python
create_table_query = """
CREATE TABLE playlist (
    id SERIAL PRIMARY KEY,
    email_user VARCHAR(200),
    title VARCHAR(250),
    creationDate DATE,
    FOREIGN KEY (email_user) REFERENCES users(email) ON DELETE CASCADE
);
"""

cur.execute(create_table_query)
conn.commit() 
```

```python
create_table_query = """
CREATE TABLE playlist_song (
    playlist_id INT,
    song_uri VARCHAR(255),
    PRIMARY KEY (playlist_id, song_uri),
    FOREIGN KEY (playlist_id) REFERENCES playlist(id) ON DELETE CASCADE,
    FOREIGN KEY (song_uri) REFERENCES song(trackURI) ON DELETE CASCADE
);
"""

cur.execute(create_table_query)
conn.commit() 
```
